# 1. Program Initiation

```python
!pip install yahoo_fin

import yahoo_fin.stock_info as yf
import pandas as pd
import time
```

```python
def get_data(ticker):
    balance_sheet = yf.get_balance_sheet(ticker)
    income_statement = yf.get_income_statement(ticker)
    cfs = yf.get_cash_flow(ticker)
    years = balance_sheet.columns
    balance_sheet.to_csv(f"{ticker} balance_sheet.csv")
    income_statement.to_csv(f"{ticker} income_statement.csv")
    cfs.to_csv(f"{ticker} cfs.csv")

```

# 2. Inputs

```python
#@title Please enter the ticker name:

if __name__ == '__main__':
  ticker_code = '1558' #@param {type:"string"}
  market = 'HK' #@param ["HK","SS"]
  ticker_name = ticker_code + '.' + market
  print(f"getting {ticker_name} data...")
  get_data(ticker_name)
  print('Done!')
```
